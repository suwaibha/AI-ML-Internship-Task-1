```python
# =====================================
# AI & ML Internship - Task 1
# Understanding Dataset & Data Types
# Dataset: Titanic
# =====================================

# ---------- Upload dataset (Colab only) ----------
try:
    from google.colab import files
    print("Upload Titanic-Dataset.csv")
    files.upload()
except:
    print("Not running in Google Colab. Skipping upload.")

# ---------- Import libraries ----------
import pandas as pd
import numpy as np
import os

# ---------- Load dataset ----------
file_name = "Titanic-Dataset.csv"

if not os.path.exists(file_name):
    raise FileNotFoundError("Titanic-Dataset.csv not found in current directory.")

df = pd.read_csv(file_name)

print("\nDataset loaded successfully!\n")

# ---------- Display first & last records ----------
print("First 5 rows:")
print(df.head())

print("\nLast 5 rows:")
print(df.tail())

# ---------- Dataset size ----------
print("\nDataset Shape (rows, columns):")
print(df.shape)

# ---------- Dataset structure ----------
print("\nDataset Info:")
print(df.info())

# ---------- Statistical summary ----------
print("\nStatistical Summary:")
print(df.describe())

# ---------- Missing values ----------
print("\nMissing Values per Column:")
print(df.isnull().sum())

# ---------- Feature type identification ----------
numerical_features = df.select_dtypes(include=["int64", "float64"]).columns
categorical_features = df.select_dtypes(include=["object"]).columns

print("\nNumerical Features:")
for col in numerical_features:
    print("-", col)

print("\nCategorical Features:")
for col in categorical_features:
    print("-", col)

# ---------- Target & input features ----------
target = "Survived"
input_features = df.drop(columns=[target])

print("\nTarget Variable:")
print(target)

print("\nInput Features:")
for col in input_features.columns:
    print("-", col)

# ---------- Target distribution (imbalance check) ----------
print("\nTarget Variable Distribution:")
print(df[target].value_counts())

# ---------- Final observations ----------
print("""
FINAL OBSERVATIONS:
1. Dataset contains 891 rows and 12 columns.
2. Target variable 'Survived' is binary.
3. Numerical, categorical, ordinal, and binary features are present.
4. Missing values exist in Age, Cabin, and Embarked columns.
5. Cabin column has excessive missing data.
6. Dataset is suitable for classical machine learning models after preprocessing.
""")
```
